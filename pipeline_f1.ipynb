```python
import fastf1
import pandas as pd
import os

cache_dir = "cache"

if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

fastf1.Cache.enable_cache(cache_dir)

start_year = 2023
end_year = 2023

results_by_year = {}
laps_by_year = {}
weather_by_year = {}
telemetry_by_year = {}
positions_by_year = {}

for year in range(start_year, end_year + 1):

    print(f"Processing season {year}")

    all_results = []
    all_laps = []
    all_weather = []
    all_telemetry = []
    all_positions = []

    schedule = fastf1.get_event_schedule(year)

    for _, event in schedule.iterrows():

        event_name = event["EventName"]
        print("Loading:", year, event_name)

        try:
            session = fastf1.get_session(year, event_name, "R")
            session.load(laps=True, telemetry=True, weather=True, messages=True)
        except Exception as e:
            print("Error loading session:", year, event_name, e)
            continue

        # RESULTADOS
        try:
            results = session.results.copy()
            results["Year"] = year
            results["Event"] = event_name
            all_results.append(results)
        except Exception as e:
            print("Error results:", year, event_name, e)

        # VUELTAS
        try:
            laps = session.laps.copy()
            laps["Year"] = year
            laps["Event"] = event_name
            all_laps.append(laps)
        except Exception as e:
            print("Error laps:", year, event_name, e)

        # CLIMA
        try:
            weather = session.weather_data.copy()
            weather["Year"] = year
            weather["Event"] = event_name
            all_weather.append(weather)
        except Exception as e:
            print("Error weather:", year, event_name, e)

        # TELEMETRÍA Y POSICIONES
        try:
            laps_source = session.laps
        except Exception as e:
            print("Skipping telemetry/positions because laps are unavailable:", year, event_name, e)
            continue

        for drv in session.drivers:
            try:
                laps_driver = laps_source.pick_drivers(drv)
            except Exception as e:
                print("Error picking driver laps:", year, event_name, drv, e)
                continue

            for _, lap in laps_driver.iterlaps():

                try:
                    tel = lap.get_car_data().add_distance()
                    tel["Driver"] = drv
                    tel["LapNumber"] = lap["LapNumber"]
                    tel["Year"] = year
                    tel["Event"] = event_name
                    all_telemetry.append(tel)
                except Exception:
                    pass

                try:
                    pos = lap.get_pos_data()
                    pos["Driver"] = drv
                    pos["LapNumber"] = lap["LapNumber"]
                    pos["Year"] = year
                    pos["Event"] = event_name
                    all_positions.append(pos)
                except Exception:
                    pass

    print("Total results:", len(all_results))
    print("Total laps:", len(all_laps))
    print("Total weather:", len(all_weather))
    print("Total telemetry:", len(all_telemetry))
    print("Total positions:", len(all_positions))

    results_by_year[year] = pd.concat(all_results, ignore_index=True) if all_results else pd.DataFrame()
    laps_by_year[year] = pd.concat(all_laps, ignore_index=True) if all_laps else pd.DataFrame()
    weather_by_year[year] = pd.concat(all_weather, ignore_index=True) if all_weather else pd.DataFrame()
    telemetry_by_year[year] = pd.concat(all_telemetry, ignore_index=True) if all_telemetry else pd.DataFrame()
    positions_by_year[year] = pd.concat(all_positions, ignore_index=True) if all_positions else pd.DataFrame()

    print(f"{year} finished and saved")
```

```python

def clean_dataframe(df):
    df = df.copy()

    # datetime -> string, pero NaT -> None
    for col in df.select_dtypes(include=["datetime64[ns]", "datetime64[ns, UTC]"]).columns:
        df[col] = df[col].apply(lambda x: x.isoformat() if pd.notnull(x) else None)

    # timedelta -> string, pero NaT -> None
    for col in df.select_dtypes(include=["timedelta64[ns]"]).columns:
        df[col] = df[col].apply(lambda x: str(x) if pd.notnull(x) else None)

    # convertir columnas object que parecen numéricas
    for col in df.columns:
        if df[col].dtype == "object":
            as_num = pd.to_numeric(df[col], errors="coerce")
            if df[col].notna().sum() > 0 and df[col].notna().sum() == as_num.notna().sum():
                df[col] = as_num

    # floats enteros -> Int64
    for col in df.select_dtypes(include=["float64", "float32"]).columns:
        non_null = df[col].dropna()
        if not non_null.empty and (non_null % 1 == 0).all():
            df[col] = df[col].astype("Int64")

    # NaN -> None
    df = df.where(pd.notnull(df), None)

    return df
```

```python
def upload_table(df, table_name):

    df = clean_dataframe(df)

    chunk = 200

    for i in range(0, len(df), chunk):

        data = df.iloc[i:i+chunk].to_dict("records")

        supabase.table(table_name).insert(data).execute()

        print(f"{table_name}: uploaded {min(i+chunk, len(df))}")
```

```python
import fastf1
import pandas as pd
import os


cache_dir = "cache"

if not os.path.exists(cache_dir):
    os.makedirs(cache_dir)

fastf1.Cache.enable_cache(cache_dir)

start_year = 2026
end_year = 2026

for year in range(start_year, end_year + 1):

    print(f"Processing season {year}")

    all_results = []
    all_laps = []
    all_weather = []
    all_telemetry = []
    all_positions = []

    schedule = fastf1.get_event_schedule(year)

    for _, event in schedule.iterrows():

        event_name = event["EventName"]

        print("Loading:", year, event_name)

        try:
            session = fastf1.get_session(year, event_name, "R")
            session.load()
        except Exception as e:
            print("Error:", year, event_name, e)
            continue

        # RESULTADOS
        results = session.results.copy()
        results["Year"] = year
        results["Event"] = event_name
        all_results.append(results)

        # VUELTAS
        laps = session.laps.copy()
        laps["Year"] = year
        laps["Event"] = event_name
        all_laps.append(laps)

        # CLIMA
        weather = session.weather_data.copy()
        weather["Year"] = year
        weather["Event"] = event_name
        all_weather.append(weather)

        # TELEMETRÍA Y POSICIONES
        for drv in session.drivers:

            laps_driver = session.laps.pick_drivers(drv)

            for _, lap in laps_driver.iterlaps():

                try:
                    tel = lap.get_car_data().add_distance()
                    tel["Driver"] = drv
                    tel["LapNumber"] = lap["LapNumber"]
                    tel["Year"] = year
                    tel["Event"] = event_name
                    all_telemetry.append(tel)
                except:
                    continue

                try:
                    pos = lap.get_pos_data()
                    pos["Driver"] = drv
                    pos["LapNumber"] = lap["LapNumber"]
                    pos["Year"] = year
                    pos["Event"] = event_name
                    all_positions.append(pos)
                except:
                    continue

    # Crear DataFrames del año
    exec(f'results_df_{year} = pd.concat(all_results, ignore_index=True)')
    exec(f'laps_df_{year} = pd.concat(all_laps, ignore_index=True)')
    exec(f'weather_df_{year} = pd.concat(all_weather, ignore_index=True)')
    exec(f'telemetry_df_{year} = pd.concat(all_telemetry, ignore_index=True)')
    exec(f'position_df_{year} = pd.concat(all_positions, ignore_index=True)')

    # Guardar cada año
    # results_df.to_parquet(f"results_{year}.parquet")
    # laps_df.to_parquet(f"laps_{year}.parquet")
    # weather_df.to_parquet(f"weather_{year}.parquet")
    exec(f'telemetry_df_{year}.to_parquet(f"telemetry_{year}.parquet")')
    exec(f'position_df_{year}.to_parquet(f"positions_{year}.parquet")')

    print(f"{year} finished and saved")
```

```python
def clean_dataframe(df):
    df = df.copy()
    
    int_cols = ["RPM", "nGear", "DRS", "LapNumber", "Year"]
    for col in int_cols:
        if col in df.columns:
            df[col] = df[col].astype("Int64")

    # ints
    int_cols = ["DriverNumber","LapNumber","Stint","TyreLife","Position","Year"]
    for col in int_cols:
        if col in df.columns:
            df[col] = df[col].astype("Int64")

    # datetime → string
    for col in df.select_dtypes(include=["datetime64[ns]"]).columns:
        df[col] = df[col].astype(str)

    # timedelta → string
    for col in df.select_dtypes(include=["timedelta64[ns]"]).columns:
        df[col] = df[col].astype(str)

    # NaN / inf
    df = df.replace([np.inf, -np.inf], None)
    df = df.replace({np.nan: None, pd.NaT: None})
    df = df.replace({"NaT": None})

    return df
```

```python
def upload_table(df, table_name):

    df = clean_dataframe(df)

    chunk = 200

    for i in range(0, len(df), chunk):

        data = df.iloc[i:i+chunk].to_dict("records")

        supabase.table(table_name).insert(data).execute()

        print(f"{table_name}: uploaded {min(i+chunk, len(df))}")
```
